```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import mean_squared_error

np.set_printoptions(threshold=np.inf)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('./datasets/winequality-red.csv')

x = df.drop('quality', axis=1)
t = pd.get_dummies(df['quality'], dtype='float32')
t[[0, 1, 2, 9, 10]] = 0.0
t = t.sort_index(axis=1)

x_train, x_test, t_train, t_test = train_test_split(x, t, train_size=.7, test_size=.3, shuffle=True)
```

```python
model = tree.DecisionTreeClassifier(max_depth=5)
model.fit(x_train, t_train)
```

```python
plt.figure(figsize=(20, 10))
_=tree.plot_tree(model)
```

```python
y_test = model.predict(x_test)
accuracy_score(t_test, y_test)
```

```python
f1_score(t_test, y_test, average=None, zero_division=0)
```

```python
df = pd.read_csv('./datasets/winequality-red.csv')

x = df.drop('quality', axis=1)
t = df['quality']

x_train, x_test, t_train, t_test = train_test_split(x, t, train_size=.7, test_size=.3, shuffle=True)
```

```python
model = tree.DecisionTreeRegressor(max_depth=5)
model.fit(x_train, t_train)
```

```python
plt.figure(figsize=(20, 10))
_=tree.plot_tree(model)
```

```python
y_test=model.predict(x_test)
mean_squared_error(y_test, t_test)
```
